## Finding Missing values


Find the missing values for all columns.


```python
#pd.set_option('display.max_rows', None)
#pd.set_option('display.max_columns', None)

missing_df = duplicate_removed_df.isnull()
missing_df.head()
```

```python
type(missing_df.columns.tolist())
```

```python
# Create three lists for writing data into a dataframe
column_list = []
false_list = []
true_list = []
for column in missing_df.columns.tolist():
    print(column)
    print(missing_df[column].value_counts())
    if len(missing_df[column].value_counts().to_list()) == 1:
           column_list.append(column)
           false_list.append(missing_df[column].value_counts().to_list()[0])
           true_list.append(0)
    else:
           column_list.append(column)
           false_list.append(missing_df[column].value_counts().to_list()[0])
           true_list.append(missing_df[column].value_counts().to_list()[1])
    print()

number_of_missing = pd.DataFrame(list(zip(column_list, false_list, true_list)), columns=['Column Name', 'Number of Non-Missing Row (False)', 'Number of Missing Row (True)'])
```

## Imputing missing values


Find the  value counts for the column WorkLoc.


```python
duplicate_removed_df[['WorkLoc']].value_counts()
```

Identify the value that is most frequent (majority) in the WorkLoc column.


```python
#make a note of the majority value here, for future reference
mode_workloc = duplicate_removed_df[['WorkLoc']].value_counts().idxmax()
```

Impute (replace) all the empty rows in the column WorkLoc with the value that you have identified as majority.


```python
import numpy as np
new_df = duplicate_removed_df
new_df['WorkLoc'] = new_df['WorkLoc'].replace(np.nan, mode_workloc)
```

After imputation there should ideally not be any empty rows in the WorkLoc column.


Verify if imputing was successful.


```python
print(new_df['WorkLoc'].isnull().sum())
print(new_df['WorkLoc'].value_counts())
print(len(new_df['WorkLoc']))
```

## Normalizing data


There are two columns in the dataset that talk about compensation.

One is "CompFreq". This column shows how often a developer is paid (Yearly, Monthly, Weekly).

The other is "CompTotal". This column talks about how much the developer is paid per Year, Month, or Week depending upon his/her "CompFreq".

This makes it difficult to compare the total compensation of the developers.

In this section you will create a new column called 'NormalizedAnnualCompensation' which contains the 'Annual Compensation' irrespective of the 'CompFreq'.

Once this column is ready, it makes comparison of salaries easy.


<hr>


List out the various categories in the column 'CompFreq'


```python
# your code goes here
print(new_df['CompFreq'].value_counts())
print(len(new_df['CompFreq']))
```

```python
# Check how these two columns look like
new_df[['CompFreq', 'CompTotal']].head()
```

```python
# Create a list for the if statement 
CompSalary = list(zip(new_df['CompFreq'], new_df['CompTotal']))
CompSalary[0][1]
```

```python
# Create a copy of the new_df to avoid making irreversible change
#new_df2 = new_df
new_df3 = new_df
# Create a new column 'NormalisedAnnualCompensation'
new_df3['NormalisedAnnualCompensation'] = ''
for i in range(len(CompSalary)):
    if CompSalary[i][0] == 'Yearly':
        new_df3.loc[i, 'NormalisedAnnualCompensation'] = CompSalary[i][1]*1     # Yearly, so keep it as it is
    elif CompSalary[i][0] == 'Monthly':
        new_df3.loc[i, 'NormalisedAnnualCompensation'] = CompSalary[i][1]*12    # Monthly, so multiply by 12
    elif CompSalary[i][0] == 'Weekly':
        new_df3.loc[i, 'NormalisedAnnualCompensation'] = CompSalary[i][1]*52    # Weekly, so multiply by 52
```
